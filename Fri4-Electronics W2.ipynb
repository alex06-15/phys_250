### 2.1 Calculation of the uncertainty of the voltage measurements with the Arduino [1 pt] 

The voltages are in 1024 discrete steps, so the uncertainty is half the step size. Thus,

$\sigma_v = \frac{0.5}{1024} \approxeq 0.0005 V$.

- Describe what range of measured voltages this uncertainty would be important [1 pt]

If V is at the smallest possible steps, it would be relevant. At 500V, like in the example, it is inconsequential.

### 3.2 Ohm’s law calculation of resistor value that should be used in the circuit. [2pts]

Since $I = 20 mA, V_\text{LED} = 2.1 V$,

Then since $5V = V_\text{LED} + V_{R} = 2.1V + V_{R}$, $V_R = 2.9V$.

Since $I$ constant, say $V_R = IR$ for the resistor, and thus $2.9V = 0.02A \cdot R$, $R = 145\Omega$.
